```python
import pandas as pd
import numpy as np
import json
import requests
import matplotlib.pyplot as plt 

#import data
#https://api.covid19india.org/raw_data6.json
response = requests.get("https://api.covid19india.org/v3/data.json")
data = json.loads(response.text)    
#print(type(data))

#extract the subset of the required data
x = data['MH']['districts']    #get the data for Maharashtra districts
#print(x)
districts_list = list(x.keys())  #districts_list contains list of all the district names in a state
#print(districts_list)
total_cases = {}
for city,c_type in x.items():
    total_cases.update({city : c_type['total']})
    
#total_cases dictionary is in the format {district: {'confirmed': x, 'deceased': y, 'recovered': z}}
#print(total_cases)

#plot the graph
N = len(total_cases)
x_loc = np.arange(start=1, stop=N+1)   #x_loc contains an array of integers from 1-N, N: number of districts
df = pd.DataFrame(total_cases)
#print(df)
df.plot(kind='bar')
```
